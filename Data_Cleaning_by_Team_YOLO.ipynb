# **Data cleaning**


```python
df_choose["Age"].replace(np.NaN,df_choose["Age"].mean())
df_choose["Flight Distance"].replace(np.NaN,df_choose["Flight Distance"].mean())
df_choose["Inflight wifi service"].replace(np.NaN,df_choose["Inflight wifi service"].mean())
df_choose["Departure/Arrival time convenient"].replace(np.NaN,df_choose["Departure/Arrival time convenient"].mean())
df_choose["Ease of Online booking"].replace(np.NaN,df_choose["Ease of Online booking"].mean())
df_choose["Gate location"].replace(np.NaN,df_choose["Gate location"].mean())
df_choose["satisfaction"].replace("satisfied",1,inplace=True)
df_choose["satisfaction"].replace("neutral or dissatisfied",0,inplace=True)
df_choose["Arrival Delay in Minutes"].replace("NA",df_choose["Arrival Delay in Minutes"].mean())
```
